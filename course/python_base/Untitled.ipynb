```python
import matplotlib.pyplot as plt
```

```python
def plot_rect(b):
    x0 = b["x_top_left"] 
    x1 = b["x_bottom_right"] 
    y0 = b["y_top_left"] 
    y1 = b["y_bottom_right"] 
    plt.plot([x0, x0, x1, x1, x0], [y0, y1, y1, y0, y0])

plt.imshow(mtrx)
block = [ 
{
  "x_top_left": 410,
  "x_bottom_right": 710,
  "y_top_left": 50,
  "y_bottom_right" : 110   
},
{
  "x_top_left": 30,
  "x_bottom_right": 410,
  "y_top_left": 150,
  "y_bottom_right" : 300   
},
{
  "x_top_left": 650,
  "x_bottom_right": 1060,
  "y_top_left": 150,
  "y_bottom_right" : 260   
},
{
  "x_top_left": 30,
  "x_bottom_right": 1030,
  "y_top_left": 350,
  "y_bottom_right" : 430   
},
]
for b in block:
    plot_rect(b)
    
```
