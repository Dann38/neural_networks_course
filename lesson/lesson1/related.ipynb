```python
import numpy as np              # библиотека для вычислений 
import matplotlib.pyplot as plt # библиотека для отрисовки графиков
```

### Математические функции

```python
print(f'np.cos(np.pi/2): {np.cos(np.pi/2)}')
print(f'np.sin(np.pi/2): {np.sin(np.pi/2)}')
print(f'np.exp(1): {np.exp(1)}')
print(f'np.sqrt(2): {np.sqrt(2)}')
```

### Работа с массивами

```python
a = [1, 2, 5, 3, 4]
```

```python
np_a = np.array(a) # массив numpy (вектор) 
np_a
```

```python
np_a.dtype # Обязательно есть тип данных
```

```python
np_a.min(), np_a.max() # Можно получить минимальный и максимальный элементы
```

```python
print(np_a) # Индексы минимального и максимального элемента
np_a.argmin(), np_a.argmax()
```

```python
# Статистики
print(np_a.mean()) # Среднее
print(np_a.std()), # СКО
print(np_a.var())  # Дисперсия
```

```python
# Часто используются сортировка
np_a.sort()
```

```python
print(np_a)
```

```python
# Иногда бывает удобно использовать сортировку по индексам
np_a = np.array(a)
index_sort = np_a.argsort()
for i in index_sort:
    print(np_a[i])
```

### Отрисовка графиков

```python
t = [0, 1, 2, 3, 4, 5]
x = [1, 1, 1, 2, 1.5, 1]
plt.plot(t, x) # простой график
```

```python
plt.plot(t, x, 'r:o') # маркеры r - цвет, : - линия, o - маркер
```

```python
plt.plot(t, x, 'r:o') # маркеры r - цвет, : - линия, o - маркер
plt.grid()
```

```python
# Иногда нужно поставить точки
plt.plot(t, x, 'r:o') # маркеры r - цвет, : - линия, o - маркер
plt.scatter(1.5, 1.5, color='b', marker='*')
plt.grid()
```

```python
# Два графика
t = np.linspace(0, 4*np.pi, 100)

plt.subplot(2, 1, 1) # Разбить график на 2 строки 1 столбец и отрисовать в 1-й области
plt.plot(t, np.cos(t), 'r')
plt.grid()


plt.subplot(2, 1, 2) # Разбить график на 2 строки 1 столбец и отрисовать в 2-й области
plt.plot(t, np.sin(t), 'b')
plt.grid()
```

```python
# Два графика
t = np.linspace(0, 4, 100)

plt.subplot(1, 2, 1) # Разбить график на 1 строку 2 столбца и отрисовать в 1-й области
plt.plot(t, np.exp(t), 'r', label='f(t) = exp(t)') # label = название графика
plt.ylabel('f(t)') # подпись оси OY
plt.xlabel('t') # подпись оси OX
plt.legend() # отобразить названия графиков
plt.grid()


plt.subplot(1, 2, 2) # Разбить график на 1 строку 2 столбца и отрисовать в 2-й области
plt.plot(t, t**2, 'b', label='f(t) = t^2')
plt.xlabel('t')
plt.legend()
plt.grid()
```
