```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def model(x, a, b, c, d):
    g = lambda t: d*b*np.cos(np.exp(t))/(1+d**2 - np.cos(b*t))
    t = np.linspace(0, x+a, 101)
    int_ = np.trapezoid(g(t), t) + c
    return int_
```

```python
a = 1 
b = 1
c = 1
d = 1
x = np.linspace(0, 1, 101)
y = [model(xi, a, b, c, d) for xi in x] 
plt.plot(x, y, 'r-')
```
